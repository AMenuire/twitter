```python
def make_clickable(link):
    # target _blank to open new window
    # extract clickable text to display for your link
    text = link.split('=')[0]
    link = link.split('=')[1]
    return f'<a target="_blank" href="{link}">{text}</a>'

```

```python
#df1 = pd.read_csv("adidas_voile.csv")
df = pd.read_csv("AF.csv")

df['text_click'] = df['text'] + " = https://twitter.com/i/web/status/" + df['tweet_id'].astype(str)
df['text_click'] = df['text_click'].apply(make_clickable)

df['sn_click'] = df['sn'] + " = https://twitter.com/" + df['sn'].astype(str)
df['sn_click'] = df['sn_click'].apply(make_clickable)




fin = np.datetime64('2020-06-01')

df['timestamp'] = pd.to_datetime(df['timestamp'])
#df = df1.append(df2)
#df = df[df['timestamp']>fin]
#df = df[df['retweeted_from'].isnull()]
#df = df[df['sn']=="jmlpyt"]
#df = df[df['text'].str.contains("arr√™tons",case=False)]
#df = df[df['retweeted_from'].isnull()]
#print(df.info())
```

## Tweet vs. retweets vs. quotes

```python
retweets_percent = df['retweeted_from'].notnull().sum()
quotes_percent = df['quoted_from'].notnull().sum()
tweets_percent = df.shape[0] - retweets_percent - quotes_percent

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Tweets', 'Retweets', 'Quotes'
sizes = [tweets_percent, retweets_percent, quotes_percent]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct=lambda p : '{:.1f}%  ({:,.0f})'.format(p,p * sum(sizes)/100))
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
sum(sizes)
#df.notnull().count()
```
